```python
import os
import json
import pandas as pd
import mysql.connector
```

```python
#extract aggregate insurance data from json file
def agg_inc_data():
    path_A="C:/Users/ruaru/Desktop/FOLDER/Phonepe pulse/pulse/data/aggregated/insurance/country/india/state/"
    agg_inc_list=os.listdir(path_A)
    column_A={'State':[], 'Year':[],'Quater':[],'Transaction_type':[], 'Transaction_count':[], 'Transaction_amount':[]}
    for state in agg_inc_list:
        path_states=path_A+state+"/"
        agg_inc_list1=os.listdir(path_states)
        for year in agg_inc_list1:
            path_year=path_states+year+"/"
            agg_inc_list2=os.listdir(path_year)
            for file in agg_inc_list2:
                path_file=path_year+file
                data=open(path_file,"r")
                D=json.load(data)
                for z in D["data"]["transactionData"]:
                    Name=z["name"]
                    count=z['paymentInstruments'][0]['count']
                    amount=z['paymentInstruments'][0]['amount']
                    column_A["State"].append(state)
                    column_A["Year"].append(year)
                    column_A["Transaction_type"].append(Name)
                    column_A["Transaction_count"].append(count)
                    column_A["Transaction_amount"].append(amount)
                    column_A["Quater"].append(int(file.strip(".json")))
    df_A=pd.DataFrame(column_A)
    df_A["State"]=df_A["State"].str.replace('andaman-&-nicobar-islands','andaman & nicobar')
    df_A["State"]=df_A["State"].str.replace("-"," ")
    df_A["State"]=df_A["State"].str.title()
    df_A["State"]=df_A["State"].str.replace("Dadra & Nagar Haveli & Daman & Diu","Dadra and Nagar Haveli and Daman and Diu")
    return (df_A)
df_A=agg_inc_data()
```
